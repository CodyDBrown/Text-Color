```python
from PIL import Image
import numpy as np
import docx
from docx import Document
from docx.shared import RGBColor
from docx.enum.section import WD_ORIENT

import PyPDF2
```

```python
frozen = open('frozen-screenplay.pdf', 'rb')
pdfReader = PyPDF2.PdfFileReader(frozen)
frozen_script = ' '
for p in range(pdfReader.numPages-1):
    
    page = pdfReader.getPage(p+1)
    text = page.extractText()
    #print(text)
    text = text.replace('\n', '')
    frozen_script += text
    

frozen_script = frozen_script.replace('Õ', "'") # Hacky bit, for some reason ' and " show up as odd characters So this changes that
frozen_script = frozen_script.replace('Ó', '"')
frozen_script = frozen_script.replace('FROZEN - J. Lee', '')
frozen_script = frozen_script.replace('  ', '') #Again kind of hacky, but it makes things look nicer
#frozen_script = frozen_script.lower()
frozen_script = frozen_script.title()
#print(frozen_script)
text_file = open("Frozen Script.txt", "w")
text_file.write(frozen_script)
text_file.close()
```

```python
#Main function
f = open('Frozen Script.txt')
lyrics = f.read()
pic = Image.open('frozen-let-it-go-photo.jpg')
pix = pic.load()
if '\n' in lyrics:
    lyrics = lyrics.replace('\n', " ")
    
doc = docx.Document()
style = doc.styles['Normal']
font = style.font
font.name = 'Noto Sans Mono CJK TC Bold'
font.size = docx.shared.Pt(6)

#Set margin size
#changing the page margins
sections = doc.sections
for section in sections:
    section.top_margin = 0
    section.bottom_margin = 0
    section.left_margin = 0
    section.right_margin = 0
    
section = doc.sections[-1]
#new_width, new_height = section.page_height, section.page_width
section.orientation = WD_ORIENT.LANDSCAPE
section.page_width = docx.shared.Inches(21.1)
section.page_height = docx.shared.Inches(28.1)

paragraph_format = doc.styles['Normal'].paragraph_format


paragraph_format.space_before = docx.shared.Pt(0)
paragraph_format.space_after = docx.shared.Pt(0)

pic_colors = pic.load()

color_list = []
for n in range(pic.height):
    for m in range(pic.width):
        color_list.append(pic_colors[m,n]) 

for i in range(len(lyrics)):
    if i%pic.width == 0:
        paraObj1 = doc.add_paragraph()
    foo = paraObj1.add_run(lyrics[i])
    font = foo.font
    font.color.rgb = RGBColor(color_list[i][0],
                              color_list[i][1], 
                              color_list[i][2])

doc.save('Frozen Test title.docx')
```

```python
red = []
green = []
blue = []
for i in range(len(lyrics)):
    red.append(color_list[i][0])
    green.append(color_list[i][1])
    blue.append(color_list[i][2])
ar = np.mean(red),
ag = np.mean(green),
ab = np.mean(blue)
print(ar, ag, ab)
```
